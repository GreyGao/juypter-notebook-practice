## Pandas索引结构

### 基本方法
- loc 用label来去定位行
- iloc 用position来去定位行

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/titanic.csv')
```

```python
df.iloc[0]
```

```python
df.iloc[0:5]
```

```python
df = df.set_index('Name')
# df
df.loc['Heikkinen, Miss. Laina', 'Age']
```

### bool类型的索引

```python
df['Age'] > 40
```

```python
df[df['Age'] > 50][:]
```

```python
df[df['Sex'] == 'female'][:5]
```

```python
df.loc[df['Sex'] == 'female', 'Age'].mean()
```

```python
(df['Age'] > 70).sum()
```
