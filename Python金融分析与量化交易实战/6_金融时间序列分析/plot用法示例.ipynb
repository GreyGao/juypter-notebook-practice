```python
# 示例 DataFrame
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1]
})
df
```

### 线形图

```python
df.plot(kind='line')
plt.show()  # 显示图表
```

### 柱状图

```python
df.plot(kind='bar')
plt.show()
```

### 直方图

```python
df['A'].plot(kind='hist')
plt.show()
```

### 散点图

```python
df.plot(kind='scatter', x='A', y='B')
plt.show()
```

### 箱形图

```python
df.plot(kind='box')
plt.show()
```

1. `figsize`
- 意义：figsize 参数用于指定图形的大小。它是一个包含两个元素的元组，分别表示图形的宽度和高度（以英寸为单位）。

- 用法：当你想要自定义图形的大小时，可以使用这个参数。

示例代码：

```python
df.plot(kind='line', figsize=(10, 5))  # 创建一个宽度为10英寸，高度为5英寸的线形图
plt.show()
```

2. `subplots`
- 意义：subplots 参数用于将 DataFrame 的每一列数据绘制到不同的子图中。当设置为 True 时，每一列将在不同的子图（subplot）中绘制。

- 用法：这个参数非常有用，当你想要对 DataFrame 中的每一列数据进行单独的可视化对比时。

示例代码：

```python
df.plot(subplots=True, figsize=(6, 8))  # 每列数据绘制到不同的子图中
plt.show()
```
